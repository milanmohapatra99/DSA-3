```python
def threeSumClosest(nums: List[int], target: int) -> int:
    # Sort the given array
    nums.sort()
    # Length of the array
    n = len(nums)
    # Closest value
    closest = nums[0] + nums[1] + nums[n - 1]
    # Loop for each element of the array
    for i in range(0, n - 2):
        # Left and right pointers
        j = i + 1
        k = n - 1
        # Loop for all other pairs
        while j < k:
            current_sum = nums[i] + nums[j] + nums[k]
            if current_sum <= target:
                j += 1
            else:
                k -= 1
            if abs(closest - target) > abs(current_sum - target):
                closest = current_sum
    return closest
```

```python
class Solution:
    def fourSum(self, nums, target):
        result = []
 
        if nums is None or len(nums) < 4:
            return result
 
        nums.sort()
 
        for i in range(len(nums) - 3):
            if i != 0 and nums[i] == nums[i - 1]:
                continue
            for j in range(i + 1, len(nums) - 2):
                if j != i + 1 and nums[j] == nums[j - 1]:
                    continue
                k = j + 1
                l = len(nums) - 1
                while k < l:
                    if nums[i] + nums[j] + nums[k] + nums[l] < target:
                        k += 1
                    elif nums[i] + nums[j] + nums[k] + nums[l] > target:
                        l -= 1
                    else:
                        t = [nums[i], nums[j], nums[k], nums[l]]
                        result.append(t)
 
                        k += 1
                        l -= 1
 
                        while k < l and nums[l] == nums[l + 1]:
                            l -= 1
 
                        while k < l and nums[k] == nums[k - 1]:
                            k += 1
        return result
```

```python
def reverse(nums, i, j):
    while i < j:
        nums[i], nums[j] = nums[j], nums[i]
        i += 1
        j -= 1


def nextPermutation(nums: List[int]):
    # Length of the array
    n = len(nums)
    # Index of the first element that is smaller than
    # the element to its right.
    index = -1
    # Loop from right to left
    for i in range(n - 1, 0, -1):
        if nums[i] > nums[i - 1]:
            index = i - 1
            break
    # Base condition
    if index == -1:
        reverse(nums, 0, n - 1)
        return
    j = n - 1
    # Again swap from right to left to find first element
    # that is greater than the above find element
    for i in range(n - 1, index, -1):
        if nums[i] > nums[index]:
            j = i
            break
    # Swap the elements
    nums[index], nums[j] = nums[j], nums[index]
    # Reverse the elements from index + 1 to the nums.length
    reverse(nums, index + 1, n - 1)
```

```python
def searchInsert(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = left + (right - left) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return left
```

```python
def plusOne(digits):
    carry = 1
    for i in range(len(digits) - 1, -1, -1):
        digits[i] += carry
        carry = digits[i] // 10
        digits[i] %= 10

    if carry == 1:
        digits.insert(0, 1)

    return digits
```

```python
def singleNumber(nums):
    result = 0
    for num in nums:
        result ^= num
    return result
```

```python
def findMissingRanges(nums, lower, upper):
    missingRanges = []
    
    # Helper function to add a range to the missingRanges list
    def addRange(start, end):
        if start == end:
            missingRanges.append(str(start))
        else:
            missingRanges.append(str(start) + "->" + str(end))
    
    # Check if the lower bound is missing
    if lower < nums[0]:
        addRange(lower, nums[0] - 1)
    
    # Check for missing numbers between adjacent elements in nums
    for i in range(len(nums) - 1):
        if nums[i + 1] > nums[i] + 1:
            addRange(nums[i] + 1, nums[i + 1] - 1)
    
    # Check if the upper bound is missing
    if upper > nums[-1]:
        addRange(nums[-1] + 1, upper)
    
    return missingRanges
```

```python
def canAttendMeetings(intervals):
    intervals.sort(key=lambda x: x[0])  # Sort the intervals based on start time
    
    for i in range(1, len(intervals)):
        if intervals[i][0] < intervals[i-1][1]:
            return False  # There is a time conflict
        
    return True  # No time conflicts found, can attend all meetings
```
